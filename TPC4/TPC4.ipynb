```python
# -------------------------
# Modelo de dados:
# Parque = [nlugares, ocupados, nome]
# Parques = [Parque, Parque, ...]
# -------------------------

def existeParque(listap, nomep):
    cond = False
    for p in listap:
        if p[2] == nomep:
            cond = True
    return cond


# 1. inserirParque(listap, p)
def inserirParque(listap, p):
    if not existeParque(listap, p[2]):
        listap.append(p)
    return listap


# 2. listar(listap)
def listar(listap):
    print("Parques disponíveis:")
    for p in listap:
        print(f"Nome: {p[2]}, Lugares: {p[0]}")


# 3. disponivel(listap, nomep, lugar)
def disponivel(listap, nomep, lugar):
    for p in listap:
        if p[2] == nomep:
            if lugar in p[1]:
                return False
            if lugar < 1 or lugar > p[0]:
                return False
            return True
    return False


# 4. estaciona(listap, nomep, lugar)
def estaciona(listap, nomep, lugar):
    for p in listap:
        if p[2] == nomep:
            if disponivel(listap, nomep, lugar):
                p[1].append(lugar)
            else:
                print("Lugar inválido ou já ocupado.")
    return listap


# 5. listardisponibilidades(listap)
def listardisponibilidades(listap):
    print("Disponibilidades:")
    for p in listap:
        livres = p[0] - len(p[1])
        print(f"{p[2]} → {livres} lugares livres")


# 6. listarP(listap, nomep)
def listarP(listap, nomep):
    for p in listap:
        if p[2] == nomep:
            print(f"Nome: {p[2]}")
            print(f"Lugares: {p[0]}")
            print(f"Ocupados: {p[1]}")


# 7. libertaLugar(listap, nomep, lugar)
def libertaLugar(listap, nomep, lugar):
    for p in listap:
        if p[2] == nomep:
            if lugar in p[1]:
                p[1].remove(lugar)
            else:
                print("Lugar não estava ocupado.")
    return listap


# 8. criaParque(listap, nomep, lugares)
def criaParque(listap, nomep, lugares):
    if not existeParque(listap, nomep):
        p = [lugares, [], nomep]
        listap.append(p)
    else:
        print("Parque já existe.")
    return listap


# 9. removeParque(listap, nomep)
def removeParque(listap, nomep):
    for p in listap:
        if p[2] == nomep:
            if len(p[1]) == 0:
                listap.remove(p)
            else:
                print("Não pode remover, parque ainda tem lugares ocupados.")
    return listap


# -------------------------
# 10. Programa principal
# -------------------------
def menu():
    parques = []

    while True:
        print("\n--- Gestão de Parques Continente ---")
        print("1 - Criar parque")
        print("2 - Listar parques")
        print("3 - Estacionar carro")
        print("4 - Libertar lugar")
        print("5 - Mostrar parque")
        print("6 - Mostrar disponibilidades")
        print("7 - Remover parque")
        print("0 - Sair")
        opcao = input("Escolha: ")

        if opcao == "1":
            nome = input("Nome do parque: ")
            lugares = int(input("Número de lugares: "))
            parques = criaParque(parques, nome, lugares)

        elif opcao == "2":
            listar(parques)

        elif opcao == "3":
            nome = input("Nome do parque: ")
            lugar = int(input("Lugar a ocupar: "))
            parques = estaciona(parques, nome, lugar)

        elif opcao == "4":
            nome = input("Nome do parque: ")
            lugar = int(input("Lugar a libertar: "))
            parques = libertaLugar(parques, nome, lugar)

        elif opcao == "5":
            nome = input("Nome do parque: ")
            listarP(parques, nome)

        elif opcao == "6":
            listardisponibilidades(parques)

        elif opcao == "7":
            nome = input("Nome do parque a remover: ")
            parques = removeParque(parques, nome)

        elif opcao == "0":
            print("A sair...")
            break

        else:
            print("Opção inválida.")


# -------------------------
# Execução
# -------------------------
menu()
```
